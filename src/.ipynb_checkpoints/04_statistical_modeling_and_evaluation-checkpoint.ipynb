# Note:
* Robustness against outliers:
Some machine learning methods are inherently more robust against outliers than others. For example, tree-based methods (like random forest or gradient boosting) are generally more robust to outliers because they split the data based on conditions, not on individual values. As for Bayesian analysis, the impact of outliers can be controlled by using appropriate likelihood functions and priors. Heavy-tailed distributions (like the Student's t-distribution) can be used to make the analysis more robust against outliers. Bayesian hierarchical models can also help by allowing parameters to vary across groups, which can accommodate outliers.

* Dealing with outliers: 
the best approach might be to perform your analysis both with and without the outliers, and report both results.

```r
library(brms)
library(dplyr)
library(tidyr)
library(pheatmap)
library(data.table)
library(ggplot2)
library(loo)
library(caret)
```

```r
# Read data
df_subset <- fread("../data/data_subset.tsv", sep = "\t")
df_subset_no_outliers <- fread("../data/data_subset_no_outliers.tsv", sep = "\t")
genes_of_interest <- as.list(unique(df_subset$GeneName))
dim(df_subset); dim(df_subset_no_outliers)
```

We do all the analysis in this notebook for both data with and without outliers.

# 1- Fitting Bayesian model 

```r
# Add a small constant to the gene_expression to make it strictly positive
df_subset$Expression <- df_subset$Expression + 0.0001

# Fit Bayesian regression model on the data with outliers
fit_with_outliers <- brm(Expression ~ GeneName + (1 | Region), 
           data = df_subset, 
           family = lognormal(),
           set_prior("lkj(2)", class = "cor"),
           prior = c(set_prior("normal(0, 2)", class = "Intercept"),
                     set_prior("cauchy(0, 2)", class = "sd")),
                     control = list(adapt_delta = 0.99, max_treedepth = 15),
           iter = 4000,
           chains = 4
          )
```

```r
# Add a small constant to the gene_expression to make it strictly positive
df_subset_no_outliers$Expression <- df_subset_no_outliers$Expression + 0.0001

# Fit Bayesian regression model on the data without outliers
fit_without_outliers <- brm(Expression ~ GeneName + (1 | Region), 
           data = df_subset_no_outliers, 
           family = lognormal(),
           set_prior("lkj(2)", class = "cor"),
           prior = c(set_prior("normal(0, 2)", class = "Intercept"),
                     set_prior("cauchy(0, 2)", class = "sd")),
                     control = list(adapt_delta = 0.99, max_treedepth = 15),
           iter = 4000,
           chains = 4
          )
```

```r
# Summary of models
summary(fit_with_outliers)
summary(fit_without_outliers)
```

In these summaries, "Estimate" gives the posterior mean, "Est.Error" gives the standard error of the posterior mean, "l-95% CI" and "u-95% CI" give the lower and upper bounds of the 95% credible interval (CI), and "Rhat" gives the potential scale reduction factor, a measure of convergence (values close to 1 indicate convergence).

Here are some insights based on the output:

* Looking at the group-level effects (~Region), the standard deviation (SD) of the intercept decreased slightly when the outliers were removed (from 0.22 to 0.20), suggesting that outliers were introducing some variability in the intercept estimates across regions.

* Looking at the population-level effects, the estimates for the effect of each gene on the expression are similar between the two models, but there are slight differences. For example, the GeneNameCACNA1G effect estimate increased slightly from 0.38 to 0.40 when outliers were removed, while for GeneNameNOS2, the effect estimate decreased slightly from -0.84 to -0.90.

* The estimate for the family-specific parameter (sigma) is also very similar between the two models.

In summary, it appears that **the presence of outliers did have some effect on the model estimates, but these effects were relatively small**. The *big picture* results of the models are quite similar, whether outliers are included or not.

### d. Checking convergence of chains

```r
# Calculate the LOO CV
loo(fit_with_outliers)

loo(fit_without_outliers)
```

**`elpd_loo`**: This stands for Expected Log Pointwise Predictive Density. This value estimates the out-of-sample expectation, a higher value is better. Here, the model with outliers has a slightly higher `elpd_loo` than the model without outliers, suggesting that it may have slightly better predictive performance.

**`p_loo`**: This value is the effective number of parameters. It provides an estimate of the model's complexity. Here, both models have similar complexity as indicated by their similar `p_loo` values.

**`looic`**: This stands for LOO Information Criterion, a measure of goodness of fit of the model. Lower values are better. Here, the model without outliers has a slightly lower `looic`, indicating it might fit the data slightly better.

**Pareto k diagnostic values**: These values help diagnose the reliability of the LOO predictions. In both models, most of the observations are good but a few (0.3%) are very bad.

In conclusion, both models have similar performance and complexity according to these metrics. However, both models have a few very problematic observations that may be affecting their performance. It might be worthwhile investigating these observations further.

# 2- Fitting linear model (analysis with machine learning)
## Data preparation

```r
# Set the seed for reproducibility
set.seed(123)

# Split the data into training and test set - with outliers
training_samples <- df_subset$Expression %>%
  createDataPartition(p = 0.8, list = FALSE)
train_data  <- df_subset[training_samples, ]
test_data <- df_subset[-training_samples, ]

# Split the data into training and test set - without outliers
training_samples_no_outliers <- df_subset_no_outliers$Expression %>%
  createDataPartition(p = 0.8, list = FALSE)
train_data_no_outliers  <- df_subset_no_outliers[training_samples, ]
test_data_no_outliers <- df_subset_no_outliers[-training_samples, ]

# Remove rows with missing values in the training datasets
train_data <- train_data[complete.cases(train_data), ]
train_data_no_outliers <- train_data_no_outliers[complete.cases(train_data_no_outliers), ]

# Remove rows with missing values in the test datasets
test_data <- test_data[complete.cases(test_data), ]
test_data_no_outliers <- test_data_no_outliers[complete.cases(test_data_no_outliers), ]
```

# Training

```r
# Fit the models with and without outliers
lm_fit_with_outliers <- train(Expression ~ GeneName + Region, 
               data = train_data, 
               method = "lm")
lm_fit_without_outliers <- train(Expression ~ GeneName + Region, 
               data = train_data_no_outliers, 
               method = "lm")
```

## Evaluation
### a. Feature importance

```r
# Calculate feature importance
importance <- varImp(lm_fit_with_outliers, scale = FALSE)
importance_no_outliers <- varImp(lm_fit_without_outliers, scale = FALSE)

# Plot feature importance with titles
plot(importance, main = "Feature Importance (with outliers)")
plot(importance_no_outliers, main = "Feature Importance (without outliers)")
```

### b. MSE

```r
# Generate predictions on the test set
test_data$pred_with_outliers <- predict(lm_fit_with_outliers, newdata = test_data)
test_data_no_outliers$pred_without_outliers <- predict(lm_fit_without_outliers, newdata = test_data_no_outliers)

# Calculate Mean Squared Error (MSE) for both models
mse_with_outliers <- mean((test_data$pred_with_outliers - test_data$Expression)^2)
mse_without_outliers <- mean((test_data_no_outliers$pred_without_outliers - test_data_no_outliers$Expression)^2)

# Print the MSE
print(paste("Mean Squared Error with outliers: ", mse_with_outliers))
print(paste("Mean Squared Error without outliers: ", mse_without_outliers))
```

### c. Comparison with Bayesian model

```r
# With outliers
# Generate predictions from the Bayesian model
posterior_preds <- posterior_predict(fit_with_outliers, newdata = test_data)

# Compute the mean of the predictions for each observation
mean_preds <- rowMeans(posterior_preds)

# Compute the MSE
bayes_mse <- mean((mean_preds - test_data$Expression)^2)

print(paste("Mean Squared Error of Bayesian model with outliers: ", bayes_mse))


# Without outliers
# Generate predictions from the Bayesian model
posterior_preds <- posterior_predict(fit_without_outliers, newdata = test_data)

# Compute the mean of the predictions for each observation
mean_preds <- rowMeans(posterior_preds)

# Compute the MSE
bayes_mse <- mean((mean_preds - test_data$Expression)^2)

print(paste("Mean Squared Error of Bayesian model wihtout outliers: ", bayes_mse))

```

Based on MSE, **linear model seems to be a better fit than Bayesian model**. However, Bayesian model provides a full posterior distribution of predictions, so it provides more information than just a point estimate. Bayesian models also include a measure of uncertainty in their predictions, which is not captured by MSE or similar metrics (i.e., RMSE, MAE, R2).
